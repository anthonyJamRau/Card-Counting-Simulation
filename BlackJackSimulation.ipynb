```python
def make_shoe(num_decks):
## function to create a shoe out of a number of decks (8)
## parameters- number of decks as an integer
## returns- a list randomized "cards"
    vals = [2,3,4,5,6,7,8,9,10,10,10,10,11]
    suits = ['spades', 'clubs', 'hearts', 'diamonds']

    decks = [0 for i in range(num_decks)]
    for i in range(0,num_decks):
        decks[i] = list(itertools.product(vals, suits))
    full_set = decks[0]
    for j in range(0,len(decks)-1):
        full_set = full_set + decks[j]
    random.shuffle(full_set)
    return full_set
```

```python
def counting_cards_HiLo(cardVal):
# function for card counting methods
# parameters- the value of a card that has been dealt as an integer
# returns- the number to add to the card count
    if cardVal in [2,3,4,5,6]:
        return 1
    elif cardVal in [7,8,9]:
        return 0
    else:
        return -1
    return

def counting_cards_Omega2(cardVal):
    if cardVal in [2,3,7]:
        return 1
    elif cardVal == 11:
        return 0
    elif cardVal in [4,5,6]:
        return 2
    elif cardVal == 9:
        return -1
    else:
        return -2
    return

def counting_cards_Wong(cardVal):
    if cardVal in [3,4,6]:
        return 1
    elif cardVal == 5:
        return 1.5
    elif cardVal in [2,7]:
        return 0.5
    elif cardVal == 8:
        return 0
    elif cardVal == 9:
        return -0.5
    else:
        return -1
    return

def counting_cards_Victor(cardVal):
    if cardVal in [2,3,4,6,7]:
        return 2
    elif cardVal in [8,11]:
        return 0
    elif cardVal == 5:
        return 3
    elif cardVal == 9:
        return -1
    else:
        return -3
    return
```

```python
def convert_aces(hand_value):
## method to change the value of an ace from 11 to 1 if needed
## parameters- the value of a hand as an integer
## returns- new hand value
    if hand_value + 11 > 21:
        return hand_value + 1    
    return hand_value + 11
```

```python
def makeBets(count,current_number_of_cards):
## method to base a bet off of the card count total
## parameters- count: the current card count of a given card counting method as integer
##             current_number: the number of cards currently in the deck as an integer
## returns- the value of a bet to be made as an integer

    ##### set the mininum bet at the table here ####
    min_bet = 20
    true_count = (count/(current_number_of_cards/52))
    num_units_to_bet = (true_count-1)
    #print(num_units_to_bet)
    if num_units_to_bet <= 1:
        bet = min_bet
    else:
        bet = num_units_to_bet * min_bet
    #print(bet)
    return bet
    
    
```

```python
#### MAIN METHOD- implementing the whole model ####
# plays 12 rounds of 8 deck shoe re shuffles and save the results of simulation in lists

# create initial lists
list_of_lists_HiLo = list()
list_of_lists_Omega2 = list()
list_of_lists_Wong = list()
list_of_lists_Victor = list()

list_of_lists_HiLoBet = list()
list_of_lists_Omega2Bet = list()
list_of_lists_WongBet = list()
list_of_lists_VictorBet = list()

list_of_lists_cards_left = list()

# 12 rounds in this for loop
for i in range(0,12):
    
    # make the shoe
    shoe = make_shoe(8)
    
    # define some reference variables for the card counting methods
    starting_number_of_cards = len(shoe)
    current_number_of_cards = starting_number_of_cards
    count_HiLo = 0
    count_Omega2 = 0
    count_Wong = 0
    count_Victor = 0
    
    net_winnings_HiLo = 0
    net_winnings_Omega2 = 0
    net_winnings_Wong = 0
    net_winnings_Victor = 0

    # store metrics here
    win_HiLo = list()
    win_Omega2 = list()
    win_Wong = list()
    win_Victor = list()
    cards_left = list()
    
    betList_HiLo = list()
    betList_Omega2 = list()
    betList_Wong = list()
    betList_Victor = list()
    

    # play each shoe
    while ((current_number_of_cards/starting_number_of_cards) > 0.25):
        #print(current_number_of_cards/starting_number_of_cards)
        dealer_hand = 0
        player_hand = 0
        bust_player = False
        bust_dealer = False
        player_twone = False

        # make bet before dealing out hands
        # bet differs between counting cards methods
        current_number_of_cards = len(shoe)
        bet_HiLo = makeBets(count_HiLo,current_number_of_cards)
        bet_Omega2 = makeBets(count_Omega2,current_number_of_cards)
        bet_Wong = makeBets(count_Wong,current_number_of_cards)
        bet_Victor = makeBets(count_Victor,current_number_of_cards)

        # play each hand
        #initialize starting values of hands
        # each time a card is dealt, count cards with a method

        dealt = shoe.pop()[0]
        count_HiLo = count_HiLo + counting_cards_HiLo(dealt)
        count_Omega2 = count_Omega2 + counting_cards_Omega2(dealt)
        count_Wong = count_Wong + counting_cards_Wong(dealt)
        count_Victor = count_Victor + counting_cards_Victor(dealt)

        player_hand = player_hand + dealt

        dealt = shoe.pop()[0]
        count_HiLo = count_HiLo + counting_cards_HiLo(dealt)
        count_Omega2 = count_Omega2 + counting_cards_Omega2(dealt)
        count_Wong = count_Wong + counting_cards_Wong(dealt)
        count_Victor = count_Victor + counting_cards_Victor(dealt)

        if dealt == 11: #ace is dealt 
            player_hand = convert_aces(player_hand)
        else:
            player_hand = player_hand + dealt

        dealt = shoe.pop()[0]
        count_HiLo = count_HiLo + counting_cards_HiLo(dealt)
        count_Omega2 = count_Omega2 + counting_cards_Omega2(dealt)
        count_Wong = count_Wong + counting_cards_Wong(dealt)
        count_Victor = count_Victor + counting_cards_Victor(dealt)

        dealer_hand = dealer_hand + dealt 

        #each iteration of this while loop might be another hit for player
        while True: 
            # look up whether to hit in the policy matrix
            decision = policy_df[[str(dealer_hand)]].loc[player_hand]
            
            if decision.values[0] == 'S':
                break # player reached a value
                
            dealt = shoe.pop()[0] # didnt stay - deal another card
            count_HiLo = count_HiLo + counting_cards_HiLo(dealt)
            count_Omega2 = count_Omega2 + counting_cards_Omega2(dealt)
            count_Wong = count_Wong + counting_cards_Wong(dealt)
            count_Victor = count_Victor + counting_cards_Victor(dealt)

            if dealt == 11: #ace is dealt so check if conversion needed
                player_hand = convert_aces(player_hand)
            else:
                player_hand = player_hand + dealt

            if player_hand > 21:
                bust_player = True
                break
            elif player_hand == 21:
                player_twone = True

                
        if not bust_player:
            if not player_twone:

                while True: # each iteration of this while loop might be another hit for dealer
                    if dealer_hand < 17:
                        dealt = shoe.pop()[0]
                        count_HiLo = count_HiLo + counting_cards_HiLo(dealt)
                        count_Omega2 = count_Omega2 + counting_cards_Omega2(dealt)
                        count_Wong = count_Wong + counting_cards_Wong(dealt)
                        count_Victor = count_Victor + counting_cards_Victor(dealt)

                        if dealt == 11:  #ace is dealt so check if conversion needed
                            dealer_hand = convert_aces(dealer_hand)
                        else:
                            dealer_hand = dealer_hand + dealt #dealer must hit
                        if dealer_hand > 21:
                            bust_dealer = True
                            break
                        elif dealer_hand >= 17:
                            break # dealer reached a value
                            
        ## check what happened and update winnings
        if player_twone:

            net_winnings_HiLo = net_winnings_HiLo + bet_HiLo
            net_winnings_Omega2 = net_winnings_Omega2 + bet_Omega2
            net_winnings_Wong = net_winnings_Wong + bet_Wong
            net_winnings_Victor = net_winnings_Victor + bet_Victor

        elif bust_player:

            net_winnings_HiLo = net_winnings_HiLo - bet_HiLo
            net_winnings_Omega2 = net_winnings_Omega2 - bet_Omega2
            net_winnings_Wong = net_winnings_Wong - bet_Wong
            net_winnings_Victor = net_winnings_Victor - bet_Victor


        elif bust_dealer:

            net_winnings_HiLo = net_winnings_HiLo + bet_HiLo
            net_winnings_Omega2 = net_winnings_Omega2 + bet_Omega2
            net_winnings_Wong = net_winnings_Wong + bet_Wong
            net_winnings_Victor = net_winnings_Victor + bet_Victor

        elif player_hand > dealer_hand:

            net_winnings_HiLo = net_winnings_HiLo + bet_HiLo
            net_winnings_Omega2 = net_winnings_Omega2 + bet_Omega2
            net_winnings_Wong = net_winnings_Wong + bet_Wong
            net_winnings_Victor = net_winnings_Victor + bet_Victor

        elif player_hand < dealer_hand:

            net_winnings_HiLo = net_winnings_HiLo - bet_HiLo
            net_winnings_Omega2 = net_winnings_Omega2 - bet_Omega2
            net_winnings_Wong = net_winnings_Wong - bet_Wong
            net_winnings_Victor = net_winnings_Victor - bet_Victor

        else:
            print()

        ## if you want to see the results of each individual hand, uncomment and run this portion -->
#         print("net winnings HiLo " + str(net_winnings_HiLo) + " current number of cards "+str(current_number_of_cards))
#         print("net winnings Omega2 " + str(net_winnings_Omega2) + " current number of cards "+str(current_number_of_cards))
#         print("net winnings Wong " + str(net_winnings_Wong) + " current number of cards "+str(current_number_of_cards))
#         print("net winnings Victor " + str(net_winnings_Victor) + " current number of cards "+str(current_number_of_cards))
#         print()



        win_HiLo.append(net_winnings_HiLo)
        win_Omega2.append(net_winnings_Omega2)
        win_Wong.append(net_winnings_Wong)
        win_Victor.append(net_winnings_Victor)
        cards_left.append(current_number_of_cards)
        
        betList_HiLo.append(bet_HiLo)
        betList_Omega2.append(bet_Omega2)
        betList_Wong.append(bet_Wong)
        betList_Victor.append(bet_Victor)
        
    # store all net winnings and bet amounts   
    list_of_lists_HiLo.append(win_HiLo)
    list_of_lists_Omega2.append(win_Omega2)
    list_of_lists_Wong.append(win_Wong)
    list_of_lists_Victor.append(win_Victor)
    list_of_lists_cards_left.append(cards_left)
    
    list_of_lists_HiLoBet.append(betList_HiLo)
    list_of_lists_Omega2Bet.append(betList_Omega2)
    list_of_lists_WongBet.append(betList_Wong)
    list_of_lists_VictorBet.append(betList_Victor)
 
```
